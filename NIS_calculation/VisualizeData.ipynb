```python
import matplotlib.pyplot as plt
from pylab import *
```

```python
nisradlist = []
nislidlist = []

with open("NISvals_laser.txt") as file:
    for line in file: 
        line = line.strip()
        nislidlist.append(line)
        
with open("NISvals_radar.txt") as file:
    for line in file: 
        line = line.strip()
        nisradlist.append(line)
```

```python
plt.figure(figsize=(8,4))
plt.plot(nisradlist)
plt.ylabel('NIS Value')
plt.xlabel('Time Step')
plt.title('Radar NIS Values')
plt.axhline(y=7.82, color='r', linestyle='-')
plt.ylim([0,15])
plt.savefig('NIS_radar')
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(nislidlist)
plt.ylabel('NIS Value')
plt.xlabel('Time Step')
plt.title('Laser NIS Values')
plt.axhline(y=5.99, color='r', linestyle='-')
plt.ylim([0,15])
plt.savefig('NIS_laser')
plt.show()
```
